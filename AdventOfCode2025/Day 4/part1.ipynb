```python


def is_roll(grid, row, col):

    if not (0<=row<len(grid)):
        return False
    
    if not (0<=col<(len(grid[0]))):
        return False
    
    return grid[row][col] == '@'


# false
print(is_roll(grid, 0,0))


# true
print(is_roll(grid, 0,8))
print(is_roll(grid, 9,8))

# out of bounds
print(is_roll(grid, -4,0))
print(is_roll(grid, 442,0))
print(is_roll(grid, 4,11))

```

```python
def count_rolls(grid, row, col):

    if grid[row][col] != '@':
        return 99
    
    result = [
        # top
        is_roll(grid, row-1, col-1),
        is_roll(grid, row-1, col),
        is_roll(grid, row-1, col+1),

        # mid
        is_roll(grid, row, col-1),
        is_roll(grid, row, col+1),

        # bottom
        is_roll(grid, row+1, col-1),
        is_roll(grid, row+1, col),
        is_roll(grid, row+1, col+1)

    ]

    return sum([1 if r else 0 for r in  result])


count_rolls(grid, 0,2)
```
