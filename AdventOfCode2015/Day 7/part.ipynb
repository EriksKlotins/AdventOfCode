```python






def parse_input(file_path):
    with open(file_path) as file:
        inp =  [line.rstrip().split(' ') for line in file]


    instructions = []


    # regularize input
    for op in inp:

        # PUT
        if len(op) == 3:
            instructions.append(['PUT', [ op[0]], op[2]])
            continue

        # NOT
        if op[0] == 'NOT':
            instructions.append(['NOT', [op[1]], op[3] ])
            continue
        # AND
        # LSHIFT
        # RSHIFT
        # OR
        if op[1] in ['AND', 'LSHIFT', 'RSHIFT', 'OR']:
            instructions.append([op[1], [op[0], op[2]],  op[4] ])
            continue
    return instructions
```

# Part I

```python
def is_inst_defined(inst, registers):
    def is_op_known(op, registers):
        # print('is_op_known', op)
        return op.isnumeric() |  (op in registers.keys())
    return all([is_op_known(k, registers) for k in inst[1]])

def enforceUINT16(var):
    return var & 0xFFFF
def valueOf(var, registers):
    """
    Takes a representation of a variable, parses is as a constant, or reads it from the registers
    and enforces 16bits lenght
    """
    if var.isnumeric():
        return enforceUINT16(int(var))
    elif var in registers.keys():
        return registers[var]
    else:
        raise Exception(f'Cannot determine value of {var}')
        return None

def valueOfList(var_list, registers):
    return [valueOf(v, registers) for v in var_list]

def perform(inst, registers):

    operands = valueOfList(inst[1], registers)
    result = None


    if inst[0] == 'PUT':
        result = operands[0]

    elif inst[0] == 'NOT':
        result = ~operands[0]
    
    elif inst[0] == 'AND':
        result = operands[0] & operands[1]
            
    elif inst[0] == 'OR':
        result = operands[0] | operands[1]
    
    elif inst[0] == 'LSHIFT':
        result = operands[0] << operands[1]

    elif inst[0] == 'RSHIFT':
        result = operands[0] >> operands[1]

    registers[inst[2]] = enforceUINT16(result)

    return registers


def find_next_defined(instructions, registers):
    for i, inst in enumerate(instructions):
        if is_inst_defined(inst, registers):
            return inst, instructions[:i]+instructions[i+1:]
    return None, []


# file_path = 'test_input.txt'

file_path = 'input.txt'
instructions = parse_input(file_path)


registers = {}
while len(instructions)>0:
    
    inst, instructions = find_next_defined(instructions, registers)
    # print(inst)
    registers = perform(inst, registers)
display(registers)

```
