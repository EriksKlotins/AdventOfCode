```python
def in_range(start, end, value):
    return start<=value<=end

def in_ranges(ranges, value):
    for (start, end) in ranges:
        if in_range(start, end,value):
            return True
    return False

```

```python
def do_overlap(range1, range2):
    return \
        (range1[0]<=range2[0]<=range1[1]) |     \
        (range1[0] <=range2[1] <=range1[1] ) |  \
        ((range2[0] <= range1[0]) & (range2[1] >= range1[1]) )


print(do_overlap([10, 14], [4,5]))
print(do_overlap( [4,5],[10, 14]))
print(do_overlap([10, 14], [12,18]))
print(do_overlap([12,18],[10, 14]))

print(do_overlap([1,20],[4, 14]))
print(do_overlap([4, 14],[1,20]))

```

```python
def merge_ranges(range1, range2):
    return [min(range1[0], range2[0]), max(range1[1], range2[1])]


print(merge_ranges([12, 18], [10, 14] ))



```
