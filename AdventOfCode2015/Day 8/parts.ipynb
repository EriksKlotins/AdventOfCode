# Part I

```python
def read_file_rb(file_path):
    with open(file_path, mode="rb") as file:
        content = file.read()

    buf = []
    tmp = []

    for b in content:
        if b == 10:
            buf.append(tmp)
            tmp = []
        else:
            tmp.append(b)
    buf.append(tmp)


    return buf
def unescape_buf(buf):
    i = 1
    result = []
    while i < len(buf[:-1]):
        if buf[i:i+2] == [92, 120]: # \x
            result.append(int('0x'+''.join([chr(a) for a in buf[i+2:i+4]]), 16))
            i += 4
        elif buf[i] == 92: # \
            result.append(buf[i+1:i+2][0])
            i+=2
        else:
            result.append(buf[i])
            i+=1
    return result
    print(result)
            

def blen(buf):
    return len(unescape_buf(buf))

def strlen(buf):
    return len(''.join([chr(a) for a in buf]))

buf = read_file_rb('input.txt')
# lines = parse_input('test_input.txt')


total = 0
for line in buf:
    total = total + strlen(line) - blen(line)

    # print(line, unescape_buf(line), strlen(line), blen(line))

# len(unescape_buf(buf[3]))

print(total)

```
