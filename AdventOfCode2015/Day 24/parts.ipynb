```python
def perms(iterable, r=None, total=20):
    # permutations('ABCD', 2) → AB AC AD BA BC BD CA CB CD DA DB DC
    # permutations(range(3)) → 012 021 102 120 201 210

    pool = tuple(iterable)
    n = len(pool)
    r = n if r is None else r
    if r > n:
        return

    indices = list(range(n))
    cycles = list(range(n, n-r, -1))
    result = tuple(pool[i] for i in indices[:r])
    if sum(result) == total:
        yield 'a', result

    while n:
        for i in reversed(range(r)):
            cycles[i] -= 1
            if cycles[i] == 0:
                indices[i:] = indices[i+1:] + indices[i:i+1]
                cycles[i] = n - i
            else:
                j = cycles[i]
                indices[i], indices[-j] = indices[-j], indices[i]
                result = tuple(pool[i] for i in indices[:r])
                # print('resukt', result)
                if sum(result)==total:
                    yield 'b', result
                break
        else:
            return

def nperms(l, total=20):
    # result = 
    for i in range(len(l)):
        # print(l)
        r = list(perms(l, r=i, total=total))
        if len(r) >0:
            print(list(r))


# list(perms([1,2,3,4,5], r=2, total=7))
nperms([1,2,3,4,5], total=7)
```
