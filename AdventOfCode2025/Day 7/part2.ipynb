```python
def read_input(filename):
    with open(filename) as file:
        lines = [list(line.rstrip(None)) for line in file]

    return lines
```

```python
grid = read_input('./input.txt')

# print(grid)
```

```python



beam_counts = [0] * len(grid[0])
beam_counts[grid[0].index('S')] = 1


for i, step in enumerate(grid):
    for j, c in enumerate(step):
        if c!='^':
            continue
        count = beam_counts[j]

        if count > 0:
            beam_counts[j - 1] += count
            beam_counts[j + 1] += count

            beam_counts[j] = 0
    # print(i, beam_counts)


print(beam_counts)
print(sum(beam_counts))
```
