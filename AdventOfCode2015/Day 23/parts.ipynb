```python
def read_input(filename):
    buf = []
    with open(filename,'r') as file:
       buf = [ line.rstrip().split(' ') for line in file]
    return buf



# code = read_input('test_input.txt')
```

```python






def do_the_magic(code, registers = {'a':0, 'b':0}):

    cur = 0
    while cur < len(code):

        inst = code[cur][0]
        args = [a.strip(',') for a in code[cur][1:]]


        # print(cur, inst, args)
        # hlf r sets register r to half its current value, then continues with the next instruction.
        if inst == 'hlf':
            registers[args[0]] //= 2
            pass

        # tpl r sets register r to triple its current value, then continues with the next instruction.
        if inst == 'tpl':
            registers[args[0]] *= 3
            pass

        # inc r increments register r, adding 1 to it, then continues with the next instruction.
        if inst == 'inc':
            registers[args[0]] += 1
            pass
        
        # jmp offset is a jump; it continues with the instruction offset away relative to itself.
        if inst == 'jmp':
            offset = int(args[0])
            cur += offset
            continue
            pass
        
        # jie r, offset is like jmp, but only jumps if register r is even ("jump if even").
        if inst == 'jie':
            offset = int(args[1])
            if registers[args[0]] % 2 == 0:
                cur += offset
                continue
            pass

            
        # jio r, offset is like jmp, but only jumps if register r is 1 ("jump if one", not odd).
        if inst == 'jio':
            offset = int(args[1])
            if registers[args[0]] == 1:
                cur += offset
                continue
            pass

        
        cur += 1
    
    
    return registers



```

# Part I

```python
code = read_input('input.txt')
# code
do_the_magic(code)
```
