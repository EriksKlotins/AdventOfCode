```python
def inc_char(char, amount=1):
    if amount == 0:
        return char, 0
    if char < 'z':
        return chr(ord(char)+1), 0
    else:
        return 'a', 1
    


print(inc_char('a'))
print(inc_char('b'))
print(inc_char('z'))
```

```python
def inc_str(string):
    carry = 1
    arr = list(string)
    i = len(arr)-1
    while (carry != 0) and (i >= 0):
        arr[i], carry = inc_char(arr[i], carry)
        i -= 1
    return ''.join(arr)





# inc_str('aaaz')

s = 'aaaaaaaa'
for i in range(1000):
    s = inc_str(s)
    print(s, end='\t')
    if i % 10 ==0:
        print('')
```

```python
def first_req(string):
    """
    Passwords must include one increasing straight of at least three letters, 
    like abc, bcd, cde, and so on, up to xyz. 
    They cannot skip letters; abd doesn't count.
    """

    for i, _ in enumerate(string[:-2]):
        a = ord(string[i])+2
        b = ord(string[i+1])+1
        c = ord(string[i+2])
        # print(a,b,c)
        if (a == b) and (b == c):
            return True
    return False


first_req('abbceffg')
```

```python
def second_req(string):
    """
    Passwords may not contain the letters i, o, or l, as these letters can be mistaken 
    for other characters and are therefore confusing.
    """

    return not any([c in string for c in ['o', 'i','l']])
    
print(second_req('aireworw'))
```

```python
def third_req(string):
    """
    Passwords must contain at least two different, 
    non-overlapping pairs of letters, like aa, bb, or zz.
    """
    
    counter = 0
    i = 0
    while (i < len(string)-1) and (counter < 2):
        if string[i] == string[i+1]:
            i+=2
            counter+=1
        else:
            i+=1
    return counter>=2

        



third_req('baaazz')
```

```python
def test_password(string):
    return all([first_req(string), second_req(string), third_req(string)])


test_password('ghjaabcc')
```

```python
def find_next(password, max_iterations=10000000):
    for i in range(max_iterations):
        password = inc_str(password)
        t = test_password(password)
        # print(p, t)
        if t:
            return password
    return None
```
