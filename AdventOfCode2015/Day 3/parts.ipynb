```python

def parse_directions(inp):
    directions = {'^':(0, -1), '>': (1, 0), '<': (-1, 0), 'v':(0, 1)}
    X, Y = [], []
    for c in inp:
        X.append( directions[c][0])
        Y.append(directions[c][1])
    

    return X, Y
    # print(cx,cy)


file_path = 'input.txt'
inp = ''
with open(file_path) as file:
    inp =  [line.rstrip() for line in file]

inp = inp.pop()

X, Y = parse_directions(inp)
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation

fig, ax = plt.subplots(1,1, figsize=(10,10))

ax.plot(X,Y)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))


def update(frame):
    global X, Y

    # for each frame, update the data stored on each artist.
    x = X[:frame]
    y = Y[:frame]
    
    return (
        ax.plot(x[:-1], y[:-1], color='gray' ),
        ax.plot(x[-2:],y[-2:], c='red', label=f'Frame = {frame}'),
        ax.legend()
        )


ani = animation.FuncAnimation(fig=fig, func=update, frames=len(X[:500]), interval=1)
ax.set_xlim(min(X), max(X))
ax.set_ylim(min(Y), max(Y))
plt.show()

ani.save(filename="path.gif", writer="pillow")



```

```python


def parse_directions(inp):
    directions = {'^':(0, -1), '>': (1, 0), '<': (-1, 0), 'v':(0, 1)}
    result = [(0,0)]
    cx, cy  = 0,0
    for c in inp:
       
        cx += directions[c][0]
        cy += directions[c][1]

        if (cx, cy) not in result:
            result.append((cx,cy))

        # print(c, directions[c], (cx, cy))
    

    return result



def merge_directions(santaDir, roboSantaDir):


# print(santaDir)
# print('---')

# print(roboSantaDir)

    result = []
    for k in santaDir+roboSantaDir:
        if k not in result:
            result.append(k)

    return result


test_inputs = {'^v':3, '^>v<':3, '^v^v^v^v^v':11}

for inp, expected in test_inputs.items():
    k = merge_directions(parse_directions(inp[::2]), parse_directions(inp[1::2]))
    print(inp, len(k), len(k) == expected)


```
