```python
file_path = 'input.txt'
inp = ''
with open(file_path) as file:
    inp =  [[int(k) for k in line.rstrip().split('x')] for line in file]
```

```python
total = 0

# inp = [[1, 1, 10]]
for i, (l, w, h) in enumerate(inp):
    a = 2*l*w + 2*w*h + 2*h*l + min(l * w, l * h, w*h )
    total += a

print(total)
```

```python
# inp = [[2,3,4]]
# inp = [[1,1,10]]
total = 0
for i, sides in enumerate(inp):
    ss = sorted(sides)
    a = ss[0]
    b = ss[1]
    c = ss[2]

    l = a*2 + b*2  + a*b*c
    total += l

total
```
