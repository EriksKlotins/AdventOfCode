```python
import numpy as np

# file_path = 'test_input.txt'
file_path = 'input.txt'
with open(file_path) as file:
    containers =  [int(line.rstrip()) for line in file]

```

```python
# In how many different ways these numbers can produce 150
containers
```

# Part I

```python
from itertools import combinations


result = sum([sum([1 for c in combinations(containers, a) if sum(c) == 150]) for a in range(3, 20)])
    

print(result)

```

# Part II

```python
# Find the minimum number of containers that can exactly fit all 150 liters
minimum = 0

a = 1
while minimum == 0:
    combs = combinations(containers, a)
    for c in combs:
        if sum(c) == 150:
            minimum = a
    a += 1

print('minimum = ', minimum)
result = sum([1 for c in combinations(containers, 4) if sum(c) == 150]) 
print('possible arrangements = ', result)
```
