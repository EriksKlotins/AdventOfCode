## Topological sorting

The below code works in the following way:

1.  Create topological order of all nodes. I think I used slightly custom method, but in essence it works in O(n) time to count how many steps from the top each node is. 

    To do it, I first convert the graph to a matrix (to_matrix_form()) and then apply topo_sort(). 


2. Given the topological order, I can calculate how deep the search needs to go. For instance, svr node is with order 0, and fft is 8, hence I can limit the search for svr->fft paths to the depth of 8

3. With this approach, I use recursive walker to count paths between svr->fft, fft->dac, and dac->out. It works fairly fast, only fft->dac took about 2 min

4. knowing how many variations there are in each segment, the final answer is their product

```python
import numpy as np




def to_matrix_form(lines):
    
    nodes= list(lines.keys()) + ['out']
    nodes_lookup = dict([(node, i) for i, node in enumerate(nodes)])
    m = np.zeros((len(nodes), len(nodes)))

    for i, node in enumerate(nodes):
        if node == 'out':
            continue

        
        for j, out in enumerate(lines[node]):
            id = nodes_lookup[out]
            m[i][id] = 1
            # print(i,id)
    return m, nodes_lookup
    # display(m)


# def print_matrix(m, lookup):
#     print('-----------')
#     print('\t', '\t'.join((lookup.keys())))
#     for i, row in enumerate(m):
#         print(list(lookup.keys())[i], "\t","\t".join([str(r) if r>0 else ' ' for r in row]))
        
#     print('-----------')
    


def topo_sort(start, matrix, nodes):
    

    stack = [start]
    buf = {start:0}

    while len(stack)>0:
        cur = stack.pop()

        # print(len(stack))

        for i, value in enumerate(matrix[cur]):
            # print(i, value)
            if (value > 0) & (i not in buf.keys()):
                # print(i, cur, list(nodes.keys())[cur], len(  buf), o)
                # print(buf)
                buf[i] = buf[cur]+1
                stack.append(i)
    return dict([(n, buf[i]) for  (n, i) in nodes.items() if i in buf.keys()])





# print(nodes.items())

print(ordering)
# topo_order(matrix, nodes)





# def topological_order(lines)
```
