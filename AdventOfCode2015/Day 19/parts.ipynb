```python
import numpy as np


def load_input(file_path):
    with open(file_path) as file:
        content =  [line.rstrip() for line in file]


    starts = []
    ends = []
    molecule = None

    for i, line in enumerate(content):
        if i == len(content)-1:
            molecule = line

        elif len(line)>0:
            start, end = line.split(' => ')
            starts.append(start )
            ends.append(end)


    # print(replacements)
    return (starts, ends), molecule

# grid

```

```python
def count_occourences(string, substr):
    start = -1
    result = 0
    while True:
        start = string.find(substr, start+1)
        if start >= 0:
            result+=1
        else:
            return result
        
        
def agg_replacements(starts, ends):
    result = dict( [(s,[]) for s in starts]  )
    for s, e in zip(starts, ends):
        result[s].append(e)

    return result



def count_replacables(molecule, starts):
    result = dict([(s, 0) for s in starts])
    
    for s in starts:
        result[s] = count_occourences(molecule, s)


    return result


```

```python
def get_max_common_len(source, target):
    result = 0
    for i in range(min(len(source), len(target))):
        if source[i] == target[i]:
            result +=1
        else:
            return result
        
        
def get_ends(state_str, states):
    """
    Generates a list of all possible outcomes given the input string and valid replacements
    """
    buf = []
    for c in state_str:
        if c in states.keys():
            buf.append(states[c] + [c])

    ye = [''.join(a) for a in product(*buf)]
    
    return ye[:-1]
```
