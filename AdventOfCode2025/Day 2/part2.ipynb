```python

# FILENAME = './test_input.txt'
FILENAME = './input.txt'

def read_input(filename):
    with open(FILENAME) as file:
        lines = [line.rstrip() for line in file]

    return lines[0]



lines = read_input(FILENAME)
# ops = [  ((1 if l[0] == 'R' else -1),  int(l[1:])  )   for l in lines] 


intervals = [ [int(f) for f in p.split('-')] for p in lines.split(',')]

print(len(intervals))
```

```python
def into_all_equal_substr(s, n = 2):
    # result = []
    ls = len(s)
    
    if ls % n != 0:
        # print('cannot make substrings',len(s) % n)
        return False
    l = ls // n
    # print('l = ',l)

    tpl = s[0:l]
    # print('tpl = ',tpl)

    for i in range(1,n):
        sub = s[i*l:(i+1)*l]
        if tpl != sub:
            return False
    
  
    return True


a = 'abcdefghijklmnopqrst'
print(into_all_equal_substr(a, n=2));
print(into_all_equal_substr(a, n=10));


print(into_all_equal_substr('1111111', n=4));

```
