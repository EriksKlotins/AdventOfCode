```python
# Linear equation system

# For example, suppose you have 8 hit points, 5 damage, and 5 armor, 

# and that the boss has 12 hit points, 7 damage, and 2 armor:

# The player deals  5-2 = 3 damage; the boss goes down to   9 hit points.
# The boss deals    7-5 = 2 damage; the player goes down to 6 hit points.

# The player deals  5-2 = 3 damage; the boss goes down to   6 hit points.
# The boss deals    7-5 = 2 damage; the player goes down to 4 hit points.

# The player deals  5-2 = 3 damage; the boss goes down to   3 hit points.
# The boss deals    7-5 = 2 damage; the player goes down to 2 hit points.

# The player deals  5-2 = 3 damage; the boss goes down to   0 hit points.

P.points/turns - B.damage + P.armor > B.points/turns - P.damage + B.armor


(P.points - B.points)/turns > B.armor - P.armor + B.damage - P.damage

P.points - B.points > (B.armor - P.armor + B.damage - P.damage)*turns




```

```python
weapons = [
    ("Dagger",        8,     4,       0),
    ("Shortsword",   10,     5,       0),
    ("Warhammer",    25,     6,       0),
    ("Longsword",    40,     7,       0),
    ("Greataxe",     74,     8,       0)
]

armor = [
    ("No armor",      0,     0,       0),
    ("Leather",      13,     0,       1),
    ("Chainmail",    31,     0,       2),
    ("Splintmail",   53,     0,       3),
    ("Bandedmail",   75,     0,       4),
    ("Platemail",   102,     0,       5)
]

rings = [
    ("No ring",       0,     0,       0),
    ("Damage +1",    25,     1,       0),
    ("Damage +2",    50,     2,       0),
    ("Damage +3",   100,     3,       0),
    ("Defense +1",   20,     0,       1),
    ("Defense +2",   40,     0,       2),
    ("Defense +3",   80,     0,       3)
]

options = []
for w in weapons:
    for a in armor:
        for r1 in rings:
            for r2 in rings:
                if (r1 != r2) or ((r1[0] == 'No ring') and (r2[0] == 'No ring')):
                    options.append((w,a,r1,r2))

                    


```

```python
def play_the_game(player_option, boss_option, player_points, boss_points):
    
    for turn in range(0,1000):

        boss_dmg = max(1, player_option[0] - boss_option[1])
        boss_points -= boss_dmg
        if boss_points <= 0:
            # print('Player won')
            return True
            break



        player_dmg = max(1, boss_option[0] - player_option[1])
        player_points -= player_dmg
        if player_points <= 0:
            # print('Boss won')
            return False
            break
       

        # print(turn, boss_dmg, player_dmg, player_points, boss_points)
        # print(turn, boss_points, player_points)


# P.points - B.points > (B.armor - P.armor + B.damage - P.damage)*turns

play_the_game((5, 5), (7, 2), player_points=8, boss_points=12)
    
```

# Parts I and II

```python
# Hit Points: 103
# Damage: 9
# Armor: 2


won_games = []
lost_games = []

for option in options:

    cost =  sum ([o[1] for o in option])
    damage = sum ([o[2] for o in option])
    armor = sum ([o[3] for o in option])
    
    result = play_the_game((damage, armor), (9,2), 100, 103)
    if result:
        won_games.append(cost)
    else:
        lost_games.append(cost)

print('Part I: ', min(won_games))
print('Part II:', max(lost_games))


```
