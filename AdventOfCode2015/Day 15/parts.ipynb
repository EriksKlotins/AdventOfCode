```python
import numpy as np
# must add up 100
amounts = [30,25,25,20]

def calculate(ingredients, amounts):

    result = []

    for i, ingredient in enumerate(ingredients):
        result.append(ingredient * amounts[i])

    return [ max(0, sum(a))  for a in np.transpose(result)]

t = calculate(data, amounts)
display(t[:-1])
print('answer is = ', prod(t[:-1]))

```
