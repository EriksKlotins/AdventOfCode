```python
def read_input(filename):
    with open(filename) as file:
        lines = [list(line.rstrip(None)) for line in file]

    return lines
```

```python
def print_grid(grid):

    for i, row in enumerate(grid):
        print(i,'\t',''.join(row))
    print('\n\n')

print_grid(grid)
```

```python
def valid_bounds(grid, start_row, start_col):

    # return (0 <= start_row < 10) & (0<=start_col<len(grid[0]) )


    return (0 <= start_row < len(grid)) & (0<=start_col<len(grid[0]) )

def beam_overlap(grid, start_row, start_col):
    return grid[start_row][start_col]  == '|'

```

# Part I

```python
grid = read_input('./input.txt')



beams = 0



def beam(grid, start_row, start_col):
    """
    Turns out, there are splitters who never get the input beam
    The puzzle is to count how many splitters receive the input.
    """

    global beams
    
    if grid[start_row][start_col] == '.':
        if valid_bounds(grid, start_row+1, start_col):        
            grid[start_row][start_col] = '|'
            if not beam_overlap(grid, start_row+1, start_col):
                beam(grid, start_row+1, start_col)
            
        

    if grid[start_row][start_col] == '^':
        
        
        beams += 1
 
        if valid_bounds(grid, start_row, start_col+1) & (not beam_overlap(grid, start_row, start_col+1)):        
            # new_beams += 1
            beam(grid, start_row,start_col+1)
        if valid_bounds(grid, start_row, start_col-1) & (not beam_overlap(grid, start_row, start_col-1)):
            # new_beams += 1
            beam(grid, start_row,start_col-1)

            

beam(grid,1,grid[0].index('S'))

print('beams = ', beams)
print_grid(grid)
print(beams)



    

```

```python

beams = []


class Beam:

    def __init__(this,start_row, start_col, history=[]):
        this.row = start_row
        this.col = start_col
        this.history = history
        this.is_done = False
    def step_forward(this):
    # here three things can happen
    # A - if unobstucted, the beam continues 
    # B - if hit a splitter, goes one way and spawns a new beam the other way and copies its memory.
    # C - this the end

        if this.is_done:
            return
        
        this.row = this.row + 1
        if valid_bounds(grid, this.row, this.col):
            if grid[this.row][this.col] == '.':
                # A route
                
                # this.history.append('|')
                pass
            if grid[this.row][ this.col] == '^': 
                # B - route
                
                beams.append(Beam(this.row, this.col + 1, this.history+['R']))
                this.col = this.col - 1 # this beam keeps left
                this.history.append('L')

        else:
            # C route
            # print('this beam is done! ', ''.join(this.history))
            this.is_done = True



```
