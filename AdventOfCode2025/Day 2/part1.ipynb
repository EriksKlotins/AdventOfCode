```python

# FILENAME = './test_input.txt'
FILENAME = './input.txt'

def read_input(filename):
    with open(FILENAME) as file:
        lines = [line.rstrip() for line in file]

    return lines[0]



lines = read_input(FILENAME)
# ops = [  ((1 if l[0] == 'R' else -1),  int(l[1:])  )   for l in lines] 


intervals = [ [int(f) for f in p.split('-')] for p in lines.split(',')]

print(len(intervals))
```

```python

# repeated pattern twice

print([ b-a for a,b in intervals ]  )
```

```python
# naive approach


total= 0
for a,b in intervals:
    # print(a,'-',b)
    for i in range(a, b+1):
        # print(i)
        n = str(i)
        if len(n) % 2 !=0:
            continue

        p1 = n[:len(n)//2]
        p2 = n[len(n)//2:]


        if p1 == p2:
            total+=i

print(total)
```
